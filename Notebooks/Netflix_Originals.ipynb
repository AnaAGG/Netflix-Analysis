```python
df= pd.read_csv("../data/netflix_titles.csv")
df.head()
```

```python
originals_df = df.merge(originals, left_on= "title", right_on="Title", how="inner")
```

```python
originals_df.columns
```

```python
originals_df.drop(['release_year', 'rating', 'duration', 'listed_in', 'description','Title', "date_added"], axis = 1, inplace=True)
```

```python
originals_df.head()
```

```python
originals_df["month"] = originals_df.apply(lambda x: sp.Cleaning(x).get_month(x["Premiere"]), axis=1)
```

```python
originals_df["year"] = originals_df.apply(lambda x: sp.Cleaning(x).get_year(x["Premiere"]), axis=1)
```

```python
originals_df.head()
```

```python
originals_df["cast"] = originals_df.apply(lambda x: sp.Cleaning(x).split_column(x["cast"]), axis=1)
```

```python
originals_df = originals_df.explode("cast")
```
