```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS


from PIL import Image
import numpy as np
```

```python
# Reading the csv file and storing it in a dataframe called df.
df = pd.read_csv("../data/netflix_clean.csv", index_col = 0)
df.head()
```

```python
# Dropping all the rows with missing values.
df.dropna(inplace=True)
```

```python
def generate_cloud_normal(column, filename):

    # Joining all the words in the description column into one string.
    text = " ".join(cat for cat in df[column])

    word_cloud = WordCloud(
        width=7000,
        height=2000,
        background_color="black",
        colormap="OrRd",
        collocations=True,
        stopwords=STOPWORDS,
        ).generate(text)

    # Showing the word cloud.
    plt.imshow(word_cloud)


    # Turning off the axis.
    plt.axis("off")

    # Showing the word cloud.
    plt.show()

    word_cloud.to_file(f'../images/world_clouds/{filename}.png')
```

```python
def generate_cloud_n(column, filename):


    mask = np.array(Image.open('../images/test.png'))

    # Joining all the words in the description column into one string.
    text = " ".join(cat for cat in df[column])

    word_cloud = WordCloud(
        width=7000,
        height=2000,
        background_color="black",
        colormap="OrRd",
        collocations=True,
        stopwords=STOPWORDS,
        mask = mask
        ).generate(text)

    # Showing the word cloud.
    plt.imshow(word_cloud)


    # Turning off the axis.
    plt.axis("off")

    # Showing the word cloud.
    plt.show()

    word_cloud.to_file(f'../images/world_clouds/{filename}.png')
```

# Description

```python
generate_cloud_normal("description", "description_normal" )
```

```python
generate_cloud_n("description", "description_N" )
```
