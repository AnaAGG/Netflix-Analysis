```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords

#TextBlob
from textblob import TextBlob

#SpaCy
import spacy
from spacy import displacy

from langdetect import detect
```

```python
df = pd.read_csv("../data/netflix_titles.csv")
df.head()
```

```python
def tokenize (string):
    try:
        tokenizer = RegexpTokenizer(r"\w+")
        tokens = tokenizer.tokenize(string)
        return tokens
    except: 
        return np.nan

    
def stop_words(lista):
    stop_words = stopwords.words("english")

    nueva_lista = []
    
    for string in lista: 
        if string.lower() not in stop_words:
            nueva_lista.append(string)
    return " ".join(nueva_lista)

def sentimientos(cancion):
    sia = SentimentIntensityAnalyzer()
    polarity = sia.polarity_scores(cancion)
    
    pol = polarity["compound"]
    return pol
```

```python

df["descr_tok"] = df["description"].apply(tokenize)
```

```python
df["descr_tok"] = df["descr_tok"].apply(stop_words)
```

```python
# Downloading the vader lexicon from the nltk library.
nltk.download("vader_lexicon")
```

```python
df["sentiment"] = df["descr_tok"].apply(sentimientos)
```
